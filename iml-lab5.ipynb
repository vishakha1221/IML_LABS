```python
'''
• to find the maximum and minimum value of a given flattened array.
Name :- Vishakha U. Chudasama
Enrollment No :- 216270307070
'''

arr = np.array([[22, 5, 6],
               [56, 76, 12],
               [21, 50, 99]])

flat_arr = arr.flatten()

max_flat = np.max(flat_arr)
min_flat = np.min(flat_arr)

print("\nOriginal Numpy array is = \n",arr)
print("\nFlattened array is = \n",flat_arr)
print("\nMaximum value in Flattened array is = ",max_flat)
print("Minimum value in Flattened array is = ",min_flat)
```

```python
'''
• to compute the mean, standard deviation, and variance of a given array along 
the second axis.
Name :- Vishakha U. Chudasama
Enrollment No :- 216270307070
'''

x = np.array([[34, 65],[12, 99],[11, 55]])

# Find the Mean of X array
mn = np.mean(x, axis=1)

# Find the Standard Deviation of X array
st_dv = np.std(x, axis=1)

#Find the Variance of X array
vr = np.var(x)

print("\nNumpy array is = \n",x)
print("\nMean of an Array = ",mn)
print("\nStandard Deviation of an array = ",st_dv)
print("\nVariance of an array = ",vr)
```
