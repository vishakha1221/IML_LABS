```python
"""Write a Pandas program to create a line plot of the opening, closing stock 
prices of given company between two specific dates.

Name :- Vishakha U. Chudasama
Enrollment No :- 216270307070
"""

import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('/kaggle/input/nifty50-stock-market-data/TATAMOTORS.csv')

start_date = '2000-01-03'  
end_date = '2000-03-13'    

df['Date'] = pd.to_datetime(df['Date'])  

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Open Price', color='green')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Close Price', color='red')

plt.title('Opening and Closing Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
