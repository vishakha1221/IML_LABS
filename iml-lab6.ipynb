```python
'''
• to convert the first column of a DataFrame as a Series.
Name :- Vishakha U. Chudasama
Enrollment No :- 216270307070
'''

# Create a DataFrame
data = {'numbers': ['1st', '2nd', '3rd', '4th'],
        'colors': ['Mango', 'Guavava', 'Apple', 'Strawberry']}
df = pd.DataFrame(data)

first_column_series = df.iloc[:, 0]

# Print the original DataFrame
print("Original DataFrame : \n", df)

# Print the extracted Series
print("\nFirst Column as a Series : \n", first_column_series)
```

```python
'''
• to create the mean and standard deviation of the data of a given Series 
Name :- Vishakha U. Chudasama
Enrollment No :- 216270307070
'''

series = pd.Series([67, 23, 67, 98, 45])

# Calculate the mean and standard deviation
mean_value = series.mean()
std_deviation = series.std()

print("Mean:", mean_value)
print("Standard Deviation:", std_deviation)
```

```python
'''
• to sort a given Series.
Name :- Vishakha U. Chudasama
Enrollment No :- 216270307070
'''

series = pd.Series([45, 87, 99, 12, 35, 73, 25])

sorted_series = series.sort_values()

print("Sorted Series : \n", sorted_series)
print(sorted_series)
```
